# SOURCE SEPARATION
Source separation is the "simple" act of separating the various composite instruments within a song into individual audio tracks. Most notably, we can extract the vocal track from a song, which proves useful for song identification.

This section simply shows the process in action: it separates a song into two components: the vocal track and the instrumental track.

We make use of a source separation library called [Demucs](https://github.com/adefossez/demucs).

```python
import torch
from google.colab import files
import os
from IPython.display import Audio, display
import shutil
```

```python
# Check the availability of the notebook's GPU. CPU is enabled by default.
# In google colab: Switch to GPU by going to Runtime > Change runtime type. select T4 GPU.
# Demucs runs on the GPU, CPU is also viable but much slower.


device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

if device == 'cpu':
    print("Warning: Running on CPU. This will be slower.")
```

## Load audio and separate the vocals

```python
# .mp3, .wav, or .flac are the recommended filetypes for our audio

print("upload audio file: ")
uploaded = files.upload()

# Get the filename
audio_file = list(uploaded.keys())[0]
print(f"\n Uploaded: {audio_file}")
```

```python
model_name = "htdemucs"  # Options: htdemucs, htdemucs_ft, mdx_extra, mdx_extra_q

print(f"Running {model_name} model...")
print("This may take 1-3 minutes depending on song length...\n")

!python -m demucs --two-stems=vocals -n {model_name} "{audio_file}"

print("\n Separation succeeded")
```

## Play the separated audio

```python
song_name = os.path.splitext(audio_file)[0]
output_dir = f"separated/{model_name}/{song_name}"

print(f"Output files saved in: {output_dir}\n")

# List all separated files
separated_files = os.listdir(output_dir)
print("Separated tracks:")
for f in separated_files:
    print(f"  • {f}")

vocals_path = os.path.join(output_dir, "vocals.wav")

print("\n Playing vocals:")
display(Audio(vocals_path))

accompaniment_path = os.path.join(output_dir, "no_vocals.wav")

print("\n Playing accompaniment (no vocals):")
display(Audio(accompaniment_path))
```
